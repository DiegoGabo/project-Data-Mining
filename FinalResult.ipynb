```python
#Importing library

import numpy as np
import pandas as pd
import datetime
```

```python
#Import result file

result = pd.read_csv('./dataset/test.csv')
```

```python
#Transform dates from '%d/%m/%Y' to datetime objects.
def transform_date(x):
    date = datetime.datetime.strptime(x, '%d/%m/%Y')
    return date

result['Date'] = result['Date'].map(transform_date)
```

```python
#Create dictionary

d = {}
```

```python
#Calculate sales of each store in each month

def estimate_sales(x):
    store = x['StoreID']
    month = x['Date'].month
    sales = x['IsOpen']
    if (store, month) in d.keys():
        d[(store, month)] += sales
    else:
        d[(store, month)] = sales
result.loc[:, ['StoreID', 'Date', 'IsOpen']].apply(estimate_sales, axis=1)
```

```python
#Create DataFrame

d1 = {'StoreID':[], 'Month':[], 'NumberOfSales':[]}

for s,m in d.keys():
    d1['StoreID'].append(s)
    d1['Month'].append(m)
    d1['NumberOfSales'].append(d[(s,m)])
    
final_result = pd.DataFrame(d1)    
```
