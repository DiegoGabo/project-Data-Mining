# Heatmaps

```python
#Importing library

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

from scipy import stats
from scipy.stats import skew
from scipy.stats import norm
from scipy.stats.stats import pearsonr
import matplotlib.pyplot as plt


%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
#Importing dataset

df = pd.read_csv("dataset/train.csv")
```

```python
fig, ax = plt.subplots(figsize=(15,15)) 
cov=df.corr(method='pearson')
sns.heatmap(cov,square=True,annot=False,cmap="Blues", ax=ax)
```

## Cluster map

```python
sns.clustermap(df.corr(), square=True, annot=True, cmap="Blues",
               linewidths=.50, figsize=(25, 25))
```

```python
import matplotlib.dates as dates
import datetime
```

```python
store1=df.loc[(df.StoreID==1000)]
def trans_ordin(day):
    return datetime.datetime.strptime(day, '%d/%m/%Y')

store1['OrdinalData'] = store1['Date'].map(trans_ordin)
```

```python
store1.head(10)
store1_no_zeros=store1.loc[(store1.NumberOfSales>0)]
```

```python
plt.figure(figsize=(40,10))
plt.plot(store1_no_zeros['OrdinalData'], store1_no_zeros['NumberOfSales'])
```
